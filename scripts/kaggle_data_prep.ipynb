# Data Preparation for Latent Mesh

Converts HuggingFace datasets to compact binary shards.
Output suitable for uploading as Kaggle Dataset.

**Set Kaggle Secrets:**
- `HF_TOKEN` — HuggingFace read token

**After running:** Upload `bin_data/` as a Kaggle Dataset named `latent-mesh-bin-data`

```python
import os, sys, subprocess
WORK = "/kaggle/working"
os.chdir(WORK)

subprocess.run(["git", "clone", "--depth", "1",
    "https://github.com/mohamed-raad/latent-mesh-diffusion"], check=True)
os.chdir("latent-mesh-diffusion")
sys.path.insert(0, "NoProp/src")

subprocess.run([sys.executable, "-m", "pip", "install", "-q",
    "torch>=2.5", "transformers", "datasets", "numpy"], check=True)
```

```python
from bin_converter import convert_hf_to_bin
from train_pipeline import PHASE_CONFIGS

cfg = PHASE_CONFIGS["phase1_250m"]
BIN_DIR = os.path.join(WORK, "bin_data")

for ds_spec in cfg.datasets:
    hf_path = ds_spec["hf_path"]
    hf_config = ds_spec.get("hf_config")
    tag = hf_path.replace("/", "_") + (f"_{hf_config}" if hf_config else "")
    bin_path = os.path.join(BIN_DIR, tag)
    print(f"Converting {tag}...")
    convert_hf_to_bin(hf_path, bin_path, hf_config=hf_config,
                      max_docs=200000, shard_size=10000,
                      max_seq_len=cfg.canvas_len)

print(f"\nAll data in {BIN_DIR}")
print("Upload this folder as a Kaggle Dataset named 'latent-mesh-bin-data'")
```
